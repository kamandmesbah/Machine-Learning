```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

a = 1
hn = [1, 1/4, 1/16]
x = np.arange(-1, 8, 0.01)

for j in range(len(hn)):
    px = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] < 0:
            px[i] = 0
        elif 0 <= x[i] <= a:
            px[i] = (1/a) * (1 - np.exp(-x[i] / hn[j]))
        else:
            px[i] = (1/a) * (np.exp(a / hn[j]) - 1) * np.exp(-x[i] / hn[j])

    if j == 0:
        plt.plot(x, px, 'r', linewidth=1.5, label='hn=1' , color = 'magenta')
    elif j == 1:
        plt.plot(x, px, 'g', linewidth=1.5, label='hn=1/4')
    elif j == 2:
        plt.plot(x, px, 'b', linewidth=1.5, label='hn=1/16' )

uniform_dist = uniform.pdf(x)
plt.plot(x, uniform_dist, 'k', linewidth=1.5, label='Uniform Dist.')

plt.legend()
plt.xlim(-0.2, 8.1)
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.show()
```
